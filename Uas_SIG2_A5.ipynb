```python
import pandas as pd
data=pd.read_excel('bireuen.xlsx')
data = pd.read_excel('bireuen.xlsx')
print("Kolom pada data dari Excel:")
print(data.columns)

# Menampilkan beberapa baris pertama dari data Excel
print(data.head())
```

```python
kota = gpd.read_file('Bireuen.geojson')
print("Kolom pada data dari GeoJSON:")
print(kota.columns)

# Menampilkan beberapa baris pertama dari data GeoJSON
print(kota.head())
```

```python
data.head()
```

```python
print("Isi kolom 'Name' pada data Excel:")
print(data['Name'].unique())

print("Isi kolom 'Name' pada data GeoJSON:")
print(kota['Name'].unique())
```

```python
# Normalisasi data di kolom 'Name'
data['Name'] = data['Name'].str.strip().str.lower()
kota['Name'] = kota['Name'].str.strip().str.lower()

# Melakukan merge pada kolom 'Name' yang sudah dinormalisasi
kota1 = kota.merge(data, on='Name')
```

```python
if kota1.empty:
    print("Dataframe hasil merge masih kosong setelah normalisasi.")
else:
    print("Dataframe hasil merge tidak kosong setelah normalisasi.")
    # Menampilkan beberapa baris pertama dari hasil merge
    print(kota1.head())
```

```python
kota1.plot(column='jumlah', legend=True, categorical=True)
```

```python
#-- konversi data ke dictionary --
dc_data=data.to_dict()
dc_data
```

```python
for i in dc_data['jumlah'].values():
  print(i)
```

```python
klasifikasi=[]
for i in dc_data['jumlah'].values():
  if i>27000:
    klasifikasi.append('Sangat Padat')
  elif i>20000 and i<=16000:
    klasifikasi.append('Padat')
  else:
    klasifikasi.append('Rendah')
```

```python
klasifikasi
```

```python
#-- Konversikan dc_data ke GeoDataFrame
gData=gpd.GeoDataFrame(dc_data)
```

```python
dc_data
```

```python
##-- Memasukkan klasifikasi ke gData --
gData['klasifikasi']=klasifikasi
```

```python
gData.head()
```

```python
kota2=kota.merge(gData, on='Name')
kota2.plot(column='klasifikasi', categorical=True, legend=True)
```
